```python
#1.
import random

# List of questions and answers
questions = [
    "What is the capital of France?",
    "What is the largest planet in our solar system?",
    "Who wrote 'Romeo and Juliet'?",
    "What is the chemical symbol for water?",
    "How many continents are there?"
]

answers = [
    "Paris",
    "Jupiter",
    "William Shakespeare",
    "H2O",
    "7"
]

# Function to play the quiz game
def play_quiz():
    # Shuffle the indices of questions
    question_indices = list(range(len(questions)))
    random.shuffle(question_indices)

    # Initialize score
    score = 0

    # Ask three randomly selected questions
    for i in range(3):
        # Get a random question
        question_index = question_indices[i]
        question = questions[question_index]
        answer = answers[question_index]

        # Ask the question and get the player's answer
        player_answer = input(f"{question} ")

        # Check if the player's answer is correct
        if player_answer.lower() == answer.lower():
            print("Correct!")
            score += 1
        else:
            print(f"Wrong! The correct answer is {answer}")

    # Print the final score
    print(f"You got {score} out of 3 questions correct.")

# Play the quiz game
play_quiz()
```

```python
#2.
def validate_phone_number(phone_number):

    phone_number = phone_number.replace("-", "")

    if len(phone_number) == 10 and phone_number.isdigit():
        return "Valid"
    elif len(phone_number) == 11 and phone_number.startswith("1") and phone_number[1:].isdigit():
        return "Valid"
    else:
        return "Invalid"

while True:
    phone_number = input("Enter a phone number: ")
    if phone_number.lower() == "quit":
        break
    print(validate_phone_number(phone_number))
```
